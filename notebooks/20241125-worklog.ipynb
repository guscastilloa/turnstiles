```python
def reader_accesos(cwd):
    accesos = pd.read_csv(cwd, sep=';')
    # file_names=os.listdir(cwd)
    # files=[]
    # for m in range(0, len(file_names)):
    #     files.append(pd.read_csv(file_names[m]))
    #Ordenar los datos de torniquetes: 
    accesos=accesos.drop(["PROGRAMA ESTUDIANTE", "edificio", "s_dia", "semana_anio", "jornada"], axis=1)
    accesos["modo_acceso"]=accesos.modo_acceso.apply(str)
    accesos=accesos[accesos.modo_acceso=="Peatonal"]
    accesos["porteria_detalle"]=accesos.porteria_detalle.str.replace("-", ";",1)
    new_cols=accesos.porteria_detalle.str.split(";", n=1, expand=True)
    accesos["entrada"]=new_cols[0]
    accesos=accesos.drop(["porteria_detalle"], axis=1)
    accesos=accesos.drop(["porteria"], axis=1)
    accesos["tipoacceso"]=(accesos.tipoacceso=="IN ").astype(int) ##In=1, Out=0 (para ahorrar espacio)
    accesos=accesos.drop_duplicates()
    accesos["fecha_completa"]=pd.to_datetime(accesos.fecha_completa, format="%Y.%m.%d %H:%M:%S")
    return accesos
```
